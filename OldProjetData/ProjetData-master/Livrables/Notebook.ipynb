###  Prsentation du choix et description de l'algorithme

<a id='Fonctionnement'></a>
####  Fonctionnement, paramtres :
Afin d'obtenir des rsultats rapidement, nous avons pris la dcision de rapidement dcouper le problme du VRP (vehicle routing problem) en plusieurs problmes de type voyageur de commerce.
Pour ce faire, nous utilisons un algorithme de k-means permettant de calculer X points dont nous allons nous servir pour regrouper les sommets du graphe en clusters. Cet algorithme prend en paramtre deux listes X et Y, les coordonnes des points du graphe ainsi que le nombre de cluster  crer. Une fois les points runis en clusters, nous n'avons plus qu' rsoudre de multiples problmes du voyageur de commerce.
Pour ce faire nous utilisons un algorithme simple qui nous permet d'avoir une solution cohrente et rapidement : un algorithme glouton. L'algorithme est appliqu sur chaque cluster, il prend en entre la liste des points du cluster ainsi que le point faisant office de dpart et d'arrive. Il calcule ensuite la distance du point de dpart  tous les autres sommets et choisi le plus proche. Ce sommet devient le sommet courant et est supprim de la liste des sommets restant. L'algorithme calcule de nouveau les distances, se dplace et fait de mme jusqu' avoir parcouru tout le sous-ensemble de points qu'est le cluster. Il revient ensuite  son point de dpart et renvoie la liste des sommet dans l'ordre dans lequel il les a parcourus.

La combinaisons de ces deux algorithmes nous permet galement de rpondre  la contrainte des 'k_camions'. En effet, nous spcifions le nombre de cluster  raliser et donc le nombre de camions disponibles pour rsoudre le graphe.

<a id='Points'></a>
#### Points positifs et ngatifs de la solution :
**Positifs :**
-	Ces deux algorithmes sont rapides  concevoir et  mettre en uvre.
-	L'excution est rapide bien que le problme soit NP-Complet puisque l'algorithme ne parcours qu'une seule fois chaque ville et n'introduit pas modification ou de mutation une fois le chemin ralis.
-	Cette implmentation nous permet de nous assurez que nous pouvons rpondre  l'appel d'offre et fournir un optimum local pour le parcours des villes.
-	Permet de configurer un nombre de camion effectuant les dplacements, permettant ainsi de minimiser la dure de retour des camions au dpt.

**Ngatifs :**
-	Ces deux algorithmes combins limites les solutions, les sommets tant "enferms" dans des clusters, l'optimal global ne se trouve peut-tre pas dans la configuration trouve lors de l'application du k-means. 
-	De la mme manire, un point lgrement loign dans un cluster ne va tre rejoint qu' la fin du tour du graph et augmenter drastiquement la distance parcourue alors qu'il aurait t prfrable de ne pas prendre la distance la plus courte mais ainsi viter un dtour. C'est le problme de l'algorithme glouton, il s'enferme dans un optimum local.
-	Le risque de s'enfermer dans un optimum local tant prsent dans chaque cluster, la probabilit de trouver l'optimum global sur la totalit du graphe est quasiment nulle.

<h3>Algorithme de clusterisation</h3>

```python
#########################################
# Algorithme K-means, cr les clusters 
#########################################

import matplotlib.pyplot as plt
import scipy as sy
import time as time
from sklearn.datasets import make_blobs
from numpy import *
from sklearn.cluster import KMeans

'''
Class that contains datas from a city
'''


class City:
        def __init__(self, x=0, y=0, name="null"):
                self.x = x
                self.y = y
                self.name = name


timeBeginning = time.process_time()

###########################
# Variables
N_cities = 150
N_camions = 10

xi = []
yi = []

cities = []

# City generation
for i in range(N_cities):
        xi.append(sy.random.randint(0,101))
        yi.append(sy.random.randint(0,101))

        
# Set starting city & delete it from the list
start = City(xi[0], yi[0])
xi.pop(0)
yi.pop(0)

coord = c_[xi,yi]

#i = 1 #Incremental cities ID


##############################
# K-means
km = KMeans(n_clusters=N_camions)
km.fit(coord)


##############################
# Uncomment to plot
cmap = plt.cm.get_cmap("hsv", km.cluster_centers_.shape[0]) #Cration de la ColorMap
fig, ax = plt.subplots(1, 1, figsize=(5, 5)) #Cration du repre
colors = [cmap(i) for i in km.fit_predict(coord)] # Cration Clustered Color Map
ax.scatter(coord[:, 0], coord[:, 1], c=colors)

        
timeEnd = time.process_time()

time = timeEnd - timeBeginning
print("Time beginning:" + str(timeBeginning))
print("Time end:" + str(timeEnd))
print("Time:" + str(time))
```

<h3>Algorithme de rsolution - Glouton</h3>

```python
#########################################
# Algorithme Glouton, rsoud dans les clusters 
#########################################

import matplotlib.pyplot as plt
import numpy as np
import math as math
import scipy as sy
import time as time


'''
Class that contains datas from a city
'''


class City:
        def __init__(self, x=0, y=0, name="null"):
                self.x = x
                self.y = y
                self.name = name

                
'''
Calculate distance between two cities
'''


def distance(currentCity, nextCity):
        distX = int(currentCity.x) - int(nextCity.x)
        distY = int(currentCity.y) - int(nextCity.y)
        dist = math.sqrt((distX**2) + (distY**2))

        return dist

    
'''
Connect two cities on the map
'''


def ConnectCities(city1, city2):
        plt.plot([int(city1.x), int(city2.x)],[int(city1.y), int(city2.y)], 'r')

        
        
# Jeu de donnes
# size = 10
# x = [10, 83, 64, 44, 8, 34, 86, 12, 70, 87, 52, 24, 29, 66, 40]
# y = [34, 90, 43, 72, 71, 28, 20, 10, 70, 54, 90, 93, 48, 16, 10]


xi = []
yi = []

for i in range(15):
        xi.append(sy.random.randint(0,101))
        yi.append(sy.random.randint(0,101))

z = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o"]
tab = np.c_[xi,yi,z]

cities = []

# Cities creation in cities tab
for item in tab:
        cities.append(City(item[0], item[1], item[2]))

start = cities[0] # City where we start
current = cities[0] # Current city

visitedCities = [] # Already visited city
visitedCities.append(start) # Starting city is stored

remainingCities = cities.copy() # Copying list of every city
remainingCities.pop(0) # Delete starting city


# Calculate distance from current to each unvisited city
# Save the shortest
# Shortest become current, popped from remaining cities and added to visited
# Do it while there still remain unvisited cities 
while len(remainingCities) > 0:
        distMin = 0
        for city in remainingCities:
                newDist = distance(current, city)
                if distMin == 0:
                        distMin = newDist
                        saved = city
                else:
                        if distMin > newDist:
                                distMin = newDist
                                saved = city
        visitedCities.append(saved)
        remainingCities.remove(saved)
        current = saved

visitedCities.append(start) # Add the start to make it a cycle

# Plot each city on the map
plt.plot(xi, yi, 'ro')

# Connect each city to the following one in the visitedCities tab
for i in range(len(visitedCities)):
        if i == len(visitedCities) -1:
                break
        ConnectCities(visitedCities[i], visitedCities[i+1])

# Display the map
plt.axis('scaled')
plt.show()
```
