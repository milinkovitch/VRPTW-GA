### 1.2.1 Représentation des contraintes
Commençons par placer les différentes contraintes

Déjà, comment tracer une droite en Python ? C'est très simple avec [matplotlib](https://matplotlib.org/3.1.0/tutorials/introductory/pyplot.html), surtout si on utilise [NumPy](https://www.courspython.com/apprendre-numpy.html) et sa bibliothèque pour manipuler des [intervales de valeurs linéaires](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html). Testons ça sur un petit exemple :

````python
import numpy as np
import matplotlib.pyplot as plt

xmax = 5
x = np.linspace(0, xmax) # on genere l'intervalle des x
y = 4 - 2*x # on genere l'intervalle des y

ymax=np.amax(y)
print(ymax)

# on trace la droite
plt.plot(x, y)

# on dessine entre 0 et les extremas de la fonction0
plt.xlim(0, xmax)
plt.ylim(0, ymax)

#on affiche
plt.show()
````

Mais ce que nous voulons faire, c'est tracer une inéquation, il faudrait donc faire ressortir la zone des valeurs respectant l'inégalité. Pour les inéquations du type $y\geq\dots$, il suffit d'utiliser [`fill_between`](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.fill_between.html) (en laissant pour le paramètre `y2` la valeur pas défaut 0). La partie respectant l'inégalité reste en blanc :

````python
y2 = 5 - 3*x
y3 = 2 - 0.8*x

ymax=max(np.amax(y), np.amax(y2), np.amax(y3))

plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)
plt.fill_between(x, y, color='lightgrey')
plt.fill_between(x, y2, color='lightgrey')
plt.fill_between(x, y3, color='lightgrey')

# on dessine entre 0,0 et les extremas des fonctions
plt.xlim(0, xmax)
plt.ylim(0, ymax)

#on affiche
plt.show()
````

Si on doit tracer une inégalité du type $y\leq\dots$, il suffit de spécifier une valeur pour le paramètre `y2` de la fonction `fill_between`. La valeur de `ymax` permettra de couvrir tout le graphique.

````python
y2 = 5 - 3*x
y3 = 2 - 0.8*x

ymax=max(np.amax(y), np.amax(y2), np.amax(y3))

plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)

plt.fill_between(x, y, ymax, color='lightgrey')
plt.fill_between(x, y2, ymax, color='lightgrey')
plt.fill_between(x, y3, ymax, color='lightgrey')

# on dessine entre 0,0 et les extremas de la fonction
plt.xlim(0, xmax)
plt.ylim(0, ymax)

#on affiche
plt.show()
````


On a donc tout ce qu'il nous faut pour afficher les inéquations de nos contraintes (les contraintes de non négativité pourront être simplement gérées en définissant les limites d'axes). Rappelons que $X_1$ est sur l'axe des abscisses et $X_2$ sur l'axe des ordonnées.<br />

Traçons les droites en rouge, et les zones d'inégalités en gris.

````python
xmax = 100
ymax= 100


x = np.arange(0, xmax) # on genere l'intervalle des x
y = 50 + x*0
y2 = 100 - x
plt.xlim(0, xmax)
plt.ylim(0, ymax)

plt.axvline(80, 0, 100, color="red")
plt.plot(x,y, color="red")
plt.plot(x,y2, color="red")

plt.fill_between(x, y, ymax, color='lightgrey')
plt.fill_between(x, y2, ymax, color='lightgrey')

#on affiche
plt.gcf().set_size_inches(18, 10) # cette fois on affiche en grand
plt.show()
````

Que représente la partie apparaissant en blanc ? Quelle est sa caractéristique mathématique ? Quelle propriété vérifie la solution optimale d'un problème linéaire en nombres réels sur cette partie ?

<em>À COMPLÉTER</em>



Un [Widget](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html) intégré [automatiquement](https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html) à notre fonction de calcul nous permettra de faire varier facilement ce paramètre et de déterminer les valeurs admissibles de $X_1$ et $X_2$ maximisant la fonction économique. Les instructions affichant le polyèdre sont déplacées dans une fonction, et on ajoute un slider permettant à l'utilisateur de générer une valeur réelle, qui est passée à la fonction en question. Il ne vous reste plus qu'à implémenter le calcul et l'affichage de la fonction économique tenant compte de cette valeur réelle :

````python
from ipywidgets import interact
import ipywidgets as widgets

def trace(n):
    xmax = 100
    ymax= 100
    x = np.arange(0, xmax) # on genere l'intervalle des x
    y = 50 + x*0

    yeco = n - x
    plt.plot(x, yeco)
    
    plt.axvline(80, 0, 100, color="red")
    plt.plot(x,y, color="red")

    plt.fill_between(x, y, ymax, color='lightgrey')
    plt.fill_between(x, yeco, ymax, color='lightgrey')

    plt.gcf().set_size_inches(18, 10) # cette fois on affiche en grand
    plt.show()
    
    
interact(trace, n=widgets.FloatSlider(min=0.0, max=100.0, step=1, value=80.0));
````

N'hésitez pas à ajuster les paramètres d'affichage (les extrema du graphique) et les steps du slider pour obtenir la réponse la plus précise possible.

## 2.2 Résolution
Nous sommes en présence d'un problème à 9 variables. La méthode de résolution graphique ne peut pas être appliquée ici, à moins de savoir faire un graphique à 9 dimensions !<br >
La méthode algébrique pourrait être utilisée, mais nous allons nous tourner vers un solveur qui calculera lui-même la solution.

Il existe de nombreux solveurs sur le marché. Si le solveur d'Excel est rapidement mis à mal, des produits orientés vers le calcul haute performance sont disponibles, tant commerciaux (Cplex, AIMMS...) qu'open source (GLPK, LP_Solve...), dont les performances et les fonctionnalités varient grandement. Ici, nous allons utiliser un framework Python nommé [PuLP](https://pythonhosted.org/PuLP/), qui permet de s'interfacer avec plusieurs solveurs (il s'interface par défaut avec [CoinMP](https://projects.coin-or.org/CoinMP), avec lequel il est livré par défaut) . L'intérêt, en dehors de pouvoir utiliser des solveurs commerciaux très performants, c'est que son usage est très intuitif.

Commençons par l'installer, puisqu'il est rarement déployé par défaut. Si vous gérez manuellement vos packages, `pip` devrait pouvoir le faire sans soucis. En revanche, si vous utilisez Anaconda, il faut passer par un channel (un dépôt de packages) spécifique :
```term
conda install -c conda-forge pulp
```

Une fois le package installé, on va s'inspirer d'[un exemple basique](https://pythonhosted.org/PuLP/CaseStudies/a_blending_problem.html) pour implémenter et résoudre le modèle :
<br />



````python
from pulp import *
import numpy as np

X1=LpVariable("X_1", 0, None)
#A COMPLETER


#probleme
prob=LpProblem("alliage", LpMinimize)

#objectif
#A COMPLETER

#contraintes
#A COMPLETER

prob.solve()
print(LpStatus[prob.status])
print("Min=", value(prob.objective))

# variables resultat
for v in prob.variables():
    print("%s=%.2f"%(v.name,v.varValue), end=', ')
````

Nous avons notre solution optimale ! On peut même pousser le raffinement jusqu'à utilier $\mathrm{\LaTeX}$ pour améliorer le rendu :

````python
from IPython.display import display, Latex

# variables resultat
result=""
for v in prob.variables():
    valeur=round(v.varValue*100, 2)
    if valeur > 0:
        result += "$"+v.name+"="+str(valeur)+"\%$, "

display(Latex(result))
````

Une remarque : PuLP accepte aussi bien des problèmes de minimisation que de maximisation, et des contraintes d'infériorité que de supériorité. Sans le savoir, l'étudiant se pose la question du [sac à dos](https://fr.wikipedia.org/wiki/Probl%C3%A8me_du_sac_%C3%A0_dos) (_knap sack problem_ en anglais, et KSP pour les intimes). On considère un sac à dos ayant une capacité maximale, et des objets ayant chacun un encombrement et une valeur. Le but est de déterminer un sous-ensemble des objets disponibles qui puisse tenir dans le sac (c'est-à-dire dont la somme des encombrements ne dépasse pas la capacité du sac) et dont la valeur totale (la somme des valeurs des objets) soit maximale. 

## 3.1 Modélisation du problème
Comment pouvons nous modéliser ce problème sous la forme d'un programme linéaire ?
<em>À COMPLÉTER</em>
<br />

Un algorithme naïf pour détermier un résultat optimal serait une approche par force brute, qui teste toutes les solutions (réalisables ou on) et conserve la meilleure solution réalisable. Dans le cas de notre étudiant, c'est une solution qui semble envisageable. Mais si l'étudiant supprime plusieurs Giga octets de jeux, et qu'il enrichit sa bibliotèque musicale, cette solution restera-t-elle viable ? Qu'est-ce qui va devenir bloquant ?

<em>À COMPLÉTER</em>


Puisque nous avons modélisé le problème sous forme linéaire, pourquoi ne pas tester l'algorithme du Simplexe dessus ? Reprenons les données issues du Smartphone de notre étudiant. Attention aux bornes de nos variables !

````python
X1=LpVariable("X_01", 0, 1)
#A COMPLETER


#probleme
prob=LpProblem("musique", LpMaximize)

#objectif
#A COMPLETER

#contraintes
#A COMPLETER

prob.solve()
print(LpStatus[prob.status])
print("Valeur totale = ", value(prob.objective))

poids=0.0
A_ub=[40, 52, 27, 68, 80, 62, 63, 53, 56, 59]
for idx, v in enumerate(prob.variables()):
    print("X%d=%d" % (idx+1, v.varValue), end=', ')
    poids+=v.varValue*A_ub[idx]
print("\npoids=%d"%poids)
````

Ça vous parait correct ? Pourtant, cela ne l'est pas ! Calculez à la main le poids de la solution, en reprenant les valeurs dans le tableau. Qu'obtenez-vous ? Pourquoi ?

<em>À COMPLÉTER</em>


Vérifions ça :

````python
poids=0.0
for idx, v in enumerate(prob.variables()):
    print("X%d=%f" % (idx+1, v.varValue), end=', ')
    poids+=v.varValue*A_ub[idx]
print("\npoids=%d"%poids)
````

Et là ça ne va plus du tout, La solution n'est pas réalisable. On a vu qu'un arrondi inférieur produisait une solution qui, elle, est réalisable. Mais avec une valeur de $X_9$ aussi proche de 1, ce n'est peut-être pas la solution optimale. Est-ce qu'un <a href="https://www.w3schools.com/python/ref_func_round.asp">arrondi au plus proche</a> marcherait mieux ?

````python
poids=0.0
for idx, v in enumerate(prob.variables()):
    print("X%d=%f" % (idx+1, v.varValue), end=', ')
    poids+=int(round(v.varValue))*A_ub[idx]
print("\npoids=%d"%poids)
````
